## Разбейте данные на выборки

Необходимо разбить данные на тренировочную, валидационную и тестовую. Будем делать это в соотношении 65:15:15

```python
features = df.drop(['is_ultra'], axis=1)
target = df['is_ultra']
features_train, features_testvalid, target_train, target_testvalid = train_test_split(
    features, target, test_size=0.35, random_state=12345)
features_test, features_valid, target_test, target_valid = train_test_split(
    features_testvalid, target_testvalid, test_size=0.5, random_state=12345)
```

<div class="alert alert-block alert-success">  
<b>Успех:</b> С разбиением все в порядке, но надеюсь, что ты не забываешь проверять полученные выборки ,  это иной раз может предостеречь тебя от глупых ошибок.
</div>  

## Исследуйте модели

***3.1 Решающее дерево***

Нужно сделать несколько вложенных циклов для перебора гиперпараметров DecisionTreeClassifier:
- criterion - функция, позволяющая оценить качество разбиения(gini и entropy, с log_loss почему-то ошибка)
- max_depth - глубина дерева
- min_samples_split - минимальное количество выборок, необходимое для разделения внутреннего узла
- min_samples_leaf - минимальное количество выборок, необходимое для нахождения в узле

```python
best_model_dt = None
best_accuracy_dt = 0
best_mean_samples_split = 0
for criterion in ['gini', 'entropy']:
    for max_depth in range(1,11):
        for min_samples_split in range(2,11):
            for min_samples_leaf in range(1,11):
                model = DecisionTreeClassifier(
                    criterion=criterion, min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf,
                    max_depth=max_depth, random_state=12345)
                model.fit(features_train, target_train)
                prediction = model.predict(features_valid)
                accuracy = accuracy_score(prediction, target_valid)
                #print(accuracy)
                if accuracy > best_accuracy_dt:
                    best_min_samples_split = min_samples_split
                    best_accuracy_dt = accuracy
                    best_model_dt = model
print('Model:', best_model_dt, '\nAccuracy: ', best_accuracy_dt)
```

Добились результата в ~0.79. Это удовлетворяет требованию, но посмотрим и другие модели. Вдруг получится еще лучше.

***3.2 Случайный лес***

Нужно сделать несколько вложенных циклов для перебора гиперпараметров RandomForestClassifier:
- n_estimators - количество деревьев в лесу
- criterion - функция, позволяющая оценить качество разбиения(gini и entropy, с log_loss почему-то ошибка)
- max_depth - глубина дерева
- min_samples_split - минимальное количество выборок, необходимое для разделения внутреннего узла
- min_samples_leaf - минимальное количество выборок, необходимое для нахождения в узле

```python
best_accuracy_rf = 0
best_model_rf = None
for est in range(1,11):
    for criterion in ['gini', 'entropy']:
        for max_depth in range(1,6):
            for min_samples_split in range(2,6):
                for min_samples_leaf in range(1,6):
                    model = RandomForestClassifier(random_state=12345, n_estimators=est, criterion=criterion,
                                                  max_depth=max_depth, min_samples_split=min_samples_split,
                                                  min_samples_leaf=min_samples_leaf)
                    model.fit(features_train, target_train)
                    prediction = model.predict(features_valid)
                    accuracy = accuracy_score(prediction, target_valid)
                    if accuracy > best_accuracy_rf:
                        best_accuracy_rf = accuracy
                        best_model_rf = model
print('Model:', best_model_rf, 'Accuracy: ', best_accuracy_rf)
```

Accuracy модели ~0.793. Немного лучше, чем у решающего дерева. Можно добиться и лучшей точности, если взять параметры больше, но время работы и так достаточно велико.

***3.3 Логистическая регрессия***

Гиперпараметры логистической регрессии:
- С - обратная величина силы регуляризации, положительное вещественное число. Будем менять от 0,001 до 5
- max_iter - максимальное число итераций


```python
best_accuracy_lg = 0
best_model_lg = None
C = np.linspace(0.001,2,100)
for c in C:
    for max_iter in [100,200,500,1000]:
        model = LogisticRegression(C=c, max_iter=max_iter, random_state=12345)
        model.fit(features_train,target_train)
        prediction = model.predict(features_valid)
        accuracy = accuracy_score(prediction, target_valid)
        if accuracy > best_accuracy_lg:
            best_accuracy_lg = accuracy
            best_model_lg = model
print('Model:', best_model_lg, 'Accuracy: ', best_accuracy_lg)
```

Accuracy оказалась заметно хуже, чем у предыдущих двух моделей.

**Выводы:**
1. Обучены три модели - Решающее дерево, Случайный лес и Логистическая регрессия.
2. Решающее дерево и Случайный лес преодолели порог точности в 0.75 на валидационной выборке
3. Лучшим с точки зрения точности оказался случайный лес - 0.793 на валидационной выборке.

<div class="alert alert-block alert-success">
<b>Успех:</b> Хорошо проделан шаг. Радует корректное использование валидационной выборки
