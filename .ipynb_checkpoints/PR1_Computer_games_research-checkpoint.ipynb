```python
# check
# –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –±–∞—Ä—ã

def pass_value_barh(df):
    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by = 'space', ascending = True)
            .plot(kind = 'barh', figsize = (19,6), rot = -5, legend = False, fontsize = 16)
            .set_title('–ü—Ä–∏–º–µ—Ä' + "\n", fontsize = 22, color = 'SteelBlue')    
        );    
    except:
        print('–ø—Ä–æ–ø—É—Å–∫–æ–≤ –Ω–µ –æ—Å—Ç–∞–ª–æ—Å—å :) –∏–ª–∏ –ø—Ä–æ–∏–∑–æ—à–ª–∞ –æ—à–∏–±–∫–∞ –≤ –ø–µ—Ä–≤–æ–π —á–∞—Å—Ç–∏ —Ñ—É–Ω–∫—Ü–∏–∏ ')
```

```python
popular_platform = (data.loc[data['year_of_release'] >= 2014]
                    .pivot_table(index='platform', values=['na_sales', 'eu_sales', 'jp_sales'], aggfunc={'sum'}).reset_index()
                   )                                                           
popular_platform.columns=['platform', 'eu_sales', 'jp_sales', 'na_sales']
for i in ['eu_sales', 'jp_sales', 'na_sales']:
    popular_platform[i] = round(popular_platform[i] / sum(popular_platform[i]),3)
display(popular_platform)
```

```python
jp_sales = popular_platform.pivot_table(index='platform', values='jp_sales').sort_values(by='jp_sales').reset_index()
jp_top5 = jp_sales['jp_sales'].iloc[4]
jp_sales['platform'] = jp_sales['platform'].where(jp_sales['jp_sales'] > jp_top5 , 'Other')
jp_sales = jp_sales.groupby('platform').sum().sort_values(by='jp_sales').reset_index()
```

```python
eu_sales = popular_platform.pivot_table(index='platform', values='eu_sales').sort_values(by='eu_sales').reset_index()
eu_top5 = eu_sales['eu_sales'].iloc[4]
eu_sales['platform'] = eu_sales['platform'].where(eu_sales['eu_sales'] > eu_top5 , 'Other')
eu_sales = eu_sales.groupby('platform').sum().sort_values(by='eu_sales').reset_index()
```

```python
na_sales = popular_platform.pivot_table(index='platform', values='na_sales').sort_values(by='na_sales').reset_index()
na_top5 = na_sales['na_sales'].iloc[4]
na_sales['platform'] = na_sales['platform'].where(na_sales['na_sales'] > na_top5 , 'Other')
na_sales = na_sales.groupby('platform').sum().sort_values(by='na_sales').reset_index()
```

## –ë–æ–Ω—É—Å

```python
data_games = pd.read_csv('/datasets/games.csv')
```

```python
data_games.columns = map(str.lower, data_games.columns)
```

```python
data_games = data_games.dropna(subset = ['year_of_release', 'name', 'genre'])
```

```python
data_games['user_score'] = data_games['user_score'].replace('tbd', np.nan).astype('float')
```

```python
data_games['rating'] = data_games['rating'].fillna('unknown')
```

```python
data_games['total_sales'] = data_games[['na_sales','eu_sales','jp_sales', 'other_sales']].sum(axis = 1)
```

```python
# check
# –∫—Ä—É–≥–∏ + –∫–∞—Ç–µ–≥–æ—Ä–∏—è –¥—Ä—É–≥–∏–µ
def graph (df, year, region, name, axes):
    
    df = df.query('year_of_release >= @year')
    
    sales = df.pivot_table(index='platform', 
                           values=region, 
                           aggfunc='sum').nlargest(5, region)
    
    sales = sales.reset_index()
    
    sales = (
            sales.append({'platform': 'Other', region: df[region].sum() 
                       - sales[region].sum()}, ignore_index= True)
         )
    
    
    sales.columns = ['platform', 'sales']
      
    labels_c=sales.platform
    colours = {'Wii':'C0', 'NES':'C1', 'GB':'C2', 'DS':'C3', 'X360':'C4', 
    'PS3':'C5', 'PS2':'C6', 'SNES':'C7', 'GBA':'C8',
               'PS4':'steelblue', '3DS':'orange', 
               'N64':'C11', 'PS':'C12', 'XB':'C13', 'PC':'C14', '2600':'C15', 'PSP':'C16', 
               'XOne':'C17',
               'WiiU':'C18', 'GC':'C19', 'GEN':'C20', 'DC':'C21', 'PSV':'C22', 
               'SAT':'C23', 'SCD':'C24', 'WS':'C25', 'NG':'C26', 
               'TG16':'C27', '3DO':'C28', 'GG':'C29', 'PCFX':'C30', 'Other':'darkred'}
   
    sales.plot(kind='pie',
               y="sales",
               
               autopct='%1.0f%%',
               wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
               textprops={'size': 'x-large'}, 
               labels= labels_c,
               colors=[colours[key] for key in labels_c],
               legend=False, 
               title = f"–ü–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å –ø–ª–∞—Ç—Ñ–æ—Ä–º –≤ {name} ", 
               ax = axes).set(ylabel='')
    
    plt.tight_layout()
```

```python
# check
# –∫—Ä—É–≥–∏ –≤ —Ä—è–¥

x_year = 2010

fig, axes = plt.subplots(1, 3, figsize = (15,6))
fig.suptitle(f'–û–±–∑–æ—Ä —Ä—ã–Ω–∫–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º (–ø–æ—Ä—Ç—Ä–µ—Ç –ø–æ–∫—É–ø–∞—Ç–µ–ª—è) —Å {x_year} –≥.', fontsize = 24, fontweight='bold')

graph(data_games, x_year, 'na_sales', '–°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ', axes[0])
graph(data_games, x_year,'eu_sales', '–ï–≤—Ä–æ–ø–µ', axes[1])
graph(data_games, x_year, 'jp_sales', '–Ø–ø–æ–Ω–∏–∏', axes[2])
```
