```python
def make_features(data, max_lag, rolling_mean_size):
    data_new = data.copy()
    data_new['month'] = data.index.month
    data_new['day'] = data.index.day
    data_new['hour'] = data.index.hour
    data_new['dayofweek'] = data.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        data_new[f'lag_{lag}'] = data['num_orders'].shift(lag)

    # < –Ω–∞–ø–∏—à–∏—Ç–µ –∫–æ–¥ –∑–¥–µ—Å—å >
    data_new['rolling_mean'] = data['num_orders'].shift(1).rolling(rolling_mean_size).mean()
    return data_new

data_new = make_features(data, 24, 7)
```

### LGBM

```python
%%time
lgbm_model = LGBMRegressor(n_jobs=-1, random_state=state)
parameters = {'num_iterations': [100, 150], 'max_depth': [12, 15]}
gridsearch_model = GridSearchCV(n_jobs=-1, cv=tss, estimator=lgbm_model, param_grid=parameters,
                                scoring='neg_root_mean_squared_error')
gridsearch_model.fit(train_features, train_target)
display(gridsearch_model.best_params_)
display(gridsearch_model.best_score_)
display(gridsearch_model.cv_results_)
```
